## CSV File with Only the Articles that cause Spikes in Political Pageviews

**Getting all political articles or respected country that displayed a spike on respected dates**

```python
import pandas as pd
import glob

allowed_dates = {
    "United States": [
        "2023-03-15", "2023-10-25", "2023-11-13",
        "2024-07-15", "2024-07-22", "2024-11-06", "2024-11-13"
    ],
    "United Kingdom": [
        "2023-05-06", "2023-11-13", "2024-07-05", "2024-11-06"
    ],
    "India": [
        "2023-03-24", "2023-05-13", "2023-12-03",
        "2024-06-04", "2024-10-10"
    ],
    "Australia": [
        "2023-05-06", "2023-10-14", "2024-11-06"
    ],
    "Canada": ["2024-11-06"]
}

csv_files = ["top_australia.csv", "top_canada.csv", "top_india.csv", "top_uk.csv", "top_united_states.csv"]

filtered_frames = []

for file in csv_files:
    df = pd.read_csv(file)

    df = df[df["label"] == "political"]

    # normlize date
    df["date"] = pd.to_datetime(df["date"]).dt.strftime("%Y-%m-%d")

    df = df[
        df["country"].isin(allowed_dates.keys()) &
        df.apply(lambda r: r["date"] in allowed_dates[r["country"]], axis=1)
    ]

    filtered_frames.append(df)

final_df = pd.concat(filtered_frames, ignore_index=True)
final_df.to_csv("spikes.csv", index=False)
```

**Getting the top 3 viewed articles of these dates**

```python
top_articles = (
    final_df
    .sort_values(["country", "date", "views"], ascending=[True, True, False])
    .groupby(["country", "date"])
    .head(3)
)
```

```python
top_articles.to_csv("top_articles_on_spike_days.csv", index=False)
```
