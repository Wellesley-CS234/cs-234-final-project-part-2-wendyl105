# Final CSV for the Streamlit

For each day from 2023 - 2024 the pageviews will be grouped by both **country code** and **label**

```python
import pandas as pd

# List your CSV files here
files = [
    "top_australia.csv",
    "top_canada.csv",
    "top_uk.csv",
    "top_india.csv",
    "top_united_states.csv"
]

dfs = []
for f in files:
    df = pd.read_csv(f)
    dfs.append(df)

full_df = pd.concat(dfs, ignore_index=True)

def normalize_label(x):
    if x == "political":
        return "political"
    elif x == "non-political":
        return "non-political"
    else:
        return "No QID"

full_df["label_group"] = full_df["label"].apply(normalize_label)

grouped = (
    full_df.groupby(["date", "country_code", "label_group"], as_index=False)["views"]
           .sum()
)

grouped = grouped.rename(columns={"label_group": "label"})

grouped.to_csv("daily_label_summary.csv", index=False)
```
